```python
Median = df['horsepower'].median()
df['horsepower']=df['horsepower'].fillna(Median)
```

```python
df['horsepower']=df['horsepower'].astype(int)
```

```python
df['num-of-doors'].value_counts()
```

```python
#Turn categorical variables into quantitative variables
df['num-of-doors'] = df['num-of-doors'].replace(['two','four'],[2,4])
```

```python
df['num-of-doors']
```

```python
def showOutliers():
    for i, col in enumerate(['num-of-doors', 'horsepower']):
        plt.subplot(2, 2, i+1)
        plt.boxplot(df[col])
        plt.title(f'Box plot of {col}')
    plt.tight_layout()
    plt.show()
showOutliers()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR
    df = df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]

df
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the data
normalized_data = scaler.fit_transform(df[df.select_dtypes(include=np.number).columns])
print("Normalized Data (Min-Max Scaling):")
print(normalized_data)
```
